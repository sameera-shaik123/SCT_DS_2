```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('train.csv')
df.head()
df.info()
df.describe()
print('stastical overview\n',df.describe())
#missing values
df.isnull().sum()
print("null values:\n",df.isnull().sum())
#handle missing values
df['Age']=df['Age'].fillna(True)
df['Cabin']=df['Cabin'].fillna(True)
df['Embarked']=df['Embarked'].fillna(True)
#checking duplicates
df.duplicated().sum()
df.drop_duplicates(inplace=True)
#univariate analysis
plt.figure(figsize=(4,4))
sns.histplot(df['Age'],kde=True,color='brown',bins=5)
plt.title('Age distribution')
plt.figure(figsize=(6,4))
df['Fare'].plot(kind='box')
plt.figure(figsize=(5,4))
sns.countplot(x='Sex',data=df)
plt.title('Gender discrimination')
#Bivariate analysis
plt.figure(figsize=(6,4))
sns.barplot(x='Sex',y='Survived',data=df,color='skyblue')
plt.title('survival by gender')
plt.figure(figsize=(10,8))
sns.barplot(x='Pclass',y='Survived',data=df,color='pink')
plt.title('survival by Ticket class')
#age distribution by survival
plt.figure(figsize=(6,4))
sns.kdeplot(df[df['Survived']==1]['Age'],label='Survived')
sns.kdeplot(df[df['Survived']==0]['Age'],label='Not survived')
#correlation matrix
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
#checking relationship
sns.pairplot(df, hue='Survived')



```
